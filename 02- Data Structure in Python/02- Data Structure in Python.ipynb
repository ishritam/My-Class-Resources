<center><h2>List</h2></center>

* A list is a collection which is ordered and changeable.

* In Python lists are written with square brackets [].

```python
a = [1,2,3,4,'firstname','lastname']
type(a)
```

```python
# Add Data Elements into list
# append(), extend(), insert()

#append()
a.append([10,20,30])
print(a)
```

```python
#extend()
a.extend([10,20,30])
print(a)
```

```python
#insert()
a.insert(4,'Hello World')
print(a)
```

## What is index
Ans: Index refers to a position within an ordered list.

```python
colors = ['red', 'green', 'blue', 'yellow', 'white', 'black']
```

###### indexing


```python
print(colors[0])
print(colors[1])
print(colors[5])
```

##### Negative indexing or Backword indexing



```python
print(colors[-1])
print(colors[-2])
print(colors[-6])
```

```python
#Erase data element from list
#del, pop(),remove()

#del
del a[4]
print(a)
```

```python
#pop()
var = a.pop(6)
print(var)

print(a)
```

```python
#remove()
a.remove(10)
print(a)
```

```python
print(num[10:0:-2])
```

```python
print(nums[::-2])
```

<center><h1> Sets</h1></center>

* Sets are un-ordered collection of unique elements.
* Sets are mutable.

```python
sets = {1,2,3,3,4,5}
print(sets)
type(sets)
```

```python
#add element in set
sets.add(10)
print(sets)
```

```python
# Functionalities of set
# union(), intersection(), difference(), symmetric_difference()

set_1 = {1,2,3,4,5,2}
set_2 = {3,4,6,7,8}
```

```python
#union
print(set_1.union(set_2))
```

```python
#intersection()
print(set_1.intersection(set_2))
```

```python
#difference()
print(set_1.difference(set_2))
```

```python
#symmetric_difference()
print(set_1.symmetric_difference(set_2))
```
